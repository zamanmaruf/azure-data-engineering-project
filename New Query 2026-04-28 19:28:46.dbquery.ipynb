```sql
-- Gold user_listening_history sample + null check on joined columns
SELECT 
  COUNT(*) as total_rows,
  SUM(CASE WHEN user_name IS NULL THEN 1 ELSE 0 END) as null_user_names,
  SUM(CASE WHEN track_name IS NULL THEN 1 ELSE 0 END) as null_track_names,
  SUM(CASE WHEN artist_name IS NULL THEN 1 ELSE 0 END) as null_artist_names,
  SUM(CASE WHEN album_name IS NULL THEN 1 ELSE 0 END) as null_album_names,
  SUM(CASE WHEN date IS NULL THEN 1 ELSE 0 END) as null_dates
FROM main.gold.gold_user_listening_history
```
